```python
def vecDpcmopt(sig,order,nCodeWords):
    var=var_model.VAR(ndf)
    results=var.fit(order)
    params=results.coefs
    err=np.zeros_like(sig)
    leng=len(sig)
    for beg in xrange(order,leng):  
        preds=results.forecast(np.array(ndf.iloc[beg-order:beg]),steps=1)
        err[beg,]=np.array(sig.iloc[beg])-preds
    codebook, idx = kmeans2(err,nCodeWords, minit='points')
    return (codebook,results)
```
